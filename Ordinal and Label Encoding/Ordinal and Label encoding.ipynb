# Ordinal Encding
- Encoding means converting categorical data into numerical data
- Encoding is important because machine learning algorithm only works with numerical data
- Ordinal encoding is used to convert input categorical data into numerical data

### Importing Libraries

```python
import numpy as np
import pandas as pd
```

### Loading Dataset

```python
df = pd.read_csv('customer.csv')
df
```

```python
df = df.iloc[:,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
X_train
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
```

```python
X_train
```

```python
oe.categories_
```

```python
X_train
```

# Label Encoding
- Encoding means converting categorical data into numerical data
- Encoding is important because machine learning algorithm only works with numerical data
- Label encoding is used to convert output categorical data into numerical data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```
